```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv

with open('csvs/nc.csv', encoding  = 'utf-8-sig') as csvfile :
    reader = csv.DictReader(csvfile)
    
    tt = []
    ZZ_norm = []
    
    for row in reader :
        tt.append(row['t'])
        ZZ_norm.append(row['Znorm'])
    
tt_arr = np.array(tt)
ZZ_norm_arr = np.array(ZZ_norm)

t = tt_arr.astype(float)
Z_norm = ZZ_norm_arr.astype(float)

plt.plot(t, Z_norm, color='green', label='Non Clotting Blood')
plt.title('Impedance Profile of a Non-Clotting Blood')
plt.xlabel('t(s)')
plt.ylabel('(Z - Zmax)/Zmax')
plt.xlim([0, 200])
plt.ylim([-0.14, 0.02])
plt.legend(loc = 'upper right')
plt.grid(True)
plt.show()
```

```python
with open('csvs/c1.csv', encoding='utf-8-sig') as csvfile:
    reader = csv.DictReader(csvfile)

    tz = []
    ZZ_clot = []

    for row in reader:
        tz.append(row['x'])
        ZZ_clot.append(row['y'])

tz_arr = np.array(tz)
ZZ_clot_arr = np.array(ZZ_clot)

ttz = tz_arr.astype(float)
Z_clot = ZZ_clot_arr.astype(float)

plt.plot(ttz, Z_clot, color='red', label='Typical Clotting Blood')
plt.title('Impedance Profile of a Typical Clotting Blood')
plt.xlabel('t(s)')
plt.ylabel('(Z - Zmax)/Zmax')
plt.xlim([0, 200])
plt.ylim([-0.14, 0.02])
plt.legend(loc = 'upper right')
plt.grid(True)
plt.show()
```
